```python
import pandas as pd

dados = pd.read_csv("Carol.Dados.csv")
dados = pd.DataFrame(dados)
dados.head()
```

# CRIANDO O GRÁFICO COM A BIBLIOTECA MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
d1= dados['Deformação(mm)']
```

```python
d2= dados['Força(N)']
```

```python
plt.figure(figsize=(15,8))
plt.plot(d1, d2)
plt.title("CP7 25% GRADE")
```

#RECRIANDO O GRÁFICO ANTERIOR COM A BIBLIOTECA SEABORN

```python
import seaborn as sns
ax = sns.lineplot(data = dados, x = "Deformação(mm)", y = "Força(N)")
ax.figure.set_size_inches(15, 8)
ax.set_xlabel("Deformação(mm)", fontsize = 16)
ax.set_ylabel("Força(N)", fontsize = 16)
ax.set_title("CP7 25% GRADE")
ax
```
